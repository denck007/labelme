```python
class Point(object):
    '''
    Mock for the QCoreQPoint object
    '''
    _x = None
    _y = None
    def __init__(self,x=None,y=None):
        if (x is not None) and (y is not None):
            self._x = float(x)
            self._y = float(y)
    def setx(self,value):
        self._x = value
    def sety(self,value):
        self._y = value
    @property
    def x(self):
        return self._x
    @property
    def y(self):
        return self._y
    def __repr__(self):
        if self._x is not None and self._y is not None:
            return "Point instance x: {:.1f} y: {:.1f}".format(self._x,self._y)
        else:
            return "Point instance not initialized"
    
class Edge:
    '''
    Hold all values to define an edge in an image
    '''

    def __init__(self,pt1,pt2,image_height,image_width):
        '''
        Given 2 points determine equation for line
        '''
        self.pt1 = pt1
        self.pt2 = pt2
        self.image_height = image_height
        self.image_width = image_width

        self.m = (pt1.y-pt2.y)/(pt1.x-pt2.x + 1e-6)
        self.b = pt1.y - self.m*pt1.x
        self.angle = np.arctan2(pt1.y-pt2.y, pt1.x-pt2.x)
        
        self.x_top = self.x_at_height(0)
        self.x_bot = self.x_at_height(self.image_height)

    def x_at_height(self,y):
        '''
        return the x location
        '''
        return (y - self.b) / (self.m + 1e-6)
    
    def get_window(self,search_width):
        '''
        Given a search_width return a bounding box [(top_left.x,top_left.y),(bottom_right.x,bottom_right.y)
        search_width is the +/- %/100 to add around the bounds of the previous edge
        '''
        x_left = min(self.x_top,self.x_bot)
        x_right = max(self.x_top,self.x_bot)

        added = self.image_width* search_width

        x_left = int(x_left - added)
        x_right = int(x_right + added)

        return [(x_left,0),(x_right,self.image_height)]
    
    def dot(self,other,normalize=False,absolute=False):
        '''
        Dot product of self and other.
        Will normalize between -1,1 if normalize == True
        Will take absolute value if absolute == True       
        '''
        v1 = [self.pt1.x - self.pt2.x, self.pt1.y - self.pt2.y]
        v2 = [other.pt1.x - other.pt2.x, other.pt1.y - other.pt2.y]
        
        dot = np.dot(v1,v2)
        if normalize:
            v1_mag = (v1[0]**2 + v1[1]**2)**0.5
            v2_mag = (v2[0]**2 + v2[1]**2)**0.5
            dot = dot / max(v1_mag,v2_mag)
        if absolute:
            dot = abs(dot)
        
        return dot
            
            
        
```

```python
#fname = images[0].split("/")[-1].strip(".jpeg")
#img = cv2.imread(os.path.join(img_dir,fname+".jpeg"))
image_name = "/home/nedenckl/railedge_cache/images/hit_13721/1002690.jpeg"
label_name = "/home/nedenckl/railedge_cache/labels_railedge/hit_13721/1002677.json"
img = cv2.imread(image_name)
with open(label_name,'r') as fp:
    label = json.load(fp)
points = label["shapes"][0]["points"]
print(points)
plt.imshow(img)
#plt.show()
for point in points:
    plt.plot(point[0],point[1],'o',markersize=5)
plt.show()

img2 = cv2.imread("/home/nedenckl/railedge_cache/images/hit_13721/1002677.jpeg")
plt.imshow(img2)
plt.show()
```

```python
height,width,_ = img.shape
pt_tl = Point(x=points[0][0],y=points[0][1])
pt_bl = Point(x=points[3][0],y=points[3][1])
print(pt_tl)
print(pt_bl)
edge_left = Edge(pt_tl,pt_bl,height,width)
b_box = edge_left.get_window(search_width=.05)
print("bounding box: {}".format(b_box))

roi = img[b_box[0][1]:b_box[1][1],b_box[0][0]:b_box[1][0],:]
plt.subplot(1,5,1)
plt.imshow(roi)

roi_eq = cv2.equalizeHist(roi[:,:,0])
plt.subplot(1,5,2)
plt.imshow(roi_eq,cmap="gray")


#                     image, threshold1, threshold2[, edges[, apertureSize[, L2gradient]]]	)
roi_canny = cv2.Canny(roi, 100,          200)
plt.subplot(1,5,3)
plt.imshow(roi_canny)

plt.subplot(1,5,4)

output = roi.copy()
#                         image,  rho, theta,           threshold[, lines[, minLineLength[, maxLineGap]]]	)
linesP = cv2.HoughLinesP(roi_canny, 1, 0.25*np.pi / 180, 50,        None,        250,         10)
if linesP is not None:
    for line in linesP:
        l = line[0]
        cv2.line(output, (l[0], l[1]), (l[2], l[3]), (0,0,255), 3, cv2.LINE_AA)
        #break
plt.imshow(output,cmap="gray")

```

```python
plt.subplot(1,5,1)

roi_gaus = cv2.blur(roi,(5,5))
plt.imshow(roi_gaus)

plt.subplot(1,5,2)
roi_sobel = cv2.Sobel(roi_gaus,cv2.CV_32F,1,0,ksize=7)
roi_sobel = np.abs(roi_sobel)
roi_sobel = roi_sobel/roi_sobel.max()*255
roi_sobel = roi_sobel.astype(np.uint8)[:,:,0]
plt.imshow(roi_sobel,cmap="gray")


roi_canny = np.zeros(roi_sobel.shape,dtype=np.uint8)
roi_canny[roi_sobel>50] = 255
#roi_canny = cv2.Canny(roi_sobel, 50,          200)
#print(roi_canny.dtype)
#print(roi_canny.shape)
plt.subplot(1,5,3)
plt.imshow(roi_canny)

plt.subplot(1,5,4)
output = roi.copy()
#                         image,  rho, theta,           threshold[, lines[, minLineLength[, maxLineGap]]]	)
linesP = cv2.HoughLinesP(roi_canny, 1, 0.1*np.pi / 180, 50,        None,        250,         10)
if linesP is not None:
    for line in linesP:
        l = line[0]
        cv2.line(output, (l[0], l[1]), (l[2], l[3]), (0,0,255), 3, cv2.LINE_AA)
plt.imshow(output)
plt.show()
```

```python
height,width,_ = img.shape
pt_tl = Point(x=points[0][0],y=points[0][1])
pt_bl = Point(x=points[3][0],y=points[3][1])

##pt_tl = Point(x=points[1][0],y=points[1][1])
#pt_bl = Point(x=points[2][0],y=points[2][1])


print(pt_tl)
print(pt_bl)
edge_left = Edge(pt_tl,pt_bl,height,width)
b_box = edge_left.get_window(search_width=.05)
print("bounding box: {}".format(b_box))

roi = img[b_box[0][1]:b_box[1][1],b_box[0][0]:b_box[1][0],:]
plt.subplot(1,6,1)
plt.imshow(roi,cmap="gray")
plt.imshow(roi)

roi_eq = cv2.equalizeHist(roi[:,:,0])
plt.subplot(1,6,2)
plt.imshow(roi_eq,cmap="gray")

plt.subplot(1,6,3)
roi_gaus = cv2.blur(roi_eq,(11,11))
plt.imshow(roi_gaus)
roi_sobel = cv2.Sobel(roi_gaus,cv2.CV_32F,1,0,ksize=7)
roi_sobel = np.abs(roi_sobel)
roi_sobel = roi_sobel/roi_sobel.max()*255
roi_sobel = roi_sobel.astype(np.uint8)[:,:]
plt.imshow(roi_sobel,cmap="gray")


plt.subplot(1,6,4)
roi_canny = np.zeros(roi_sobel.shape,dtype=np.uint8)
roi_canny[roi_sobel>75] = 255
plt.imshow(roi_canny)

plt.subplot(1,6,5)
output = roi.copy()
#                         image,  rho, theta,           threshold[, lines[, minLineLength[, maxLineGap]]]	)
linesP = cv2.HoughLinesP(roi_canny, 1, 0.25*np.pi / 180, 50,        None,        250,         50)

edge_ref = Edge(pt_tl,pt_bl,1200,1600)
#print("ref: dx: {:.0f} dy: {:.0f} angle: {:.3f}".format(ref[0],ref[1],ref_angle))

if linesP is not None:
    best_edge = None
    best_x_offset = 2*width
    for line in linesP:
        l = line[0]
        pt1 = Point(l[0],l[1])
        pt2 = Point(l[2],l[3])
        prop = Edge(pt1,pt2,1200,1600)
        if best_edge is None:
            best_edge = prop
        #print("ref: {:6.3f} prop: {:6.3f} delta: {:.6f}".format(edge_ref.angle,prop.angle,abs(edge_ref.angle-prop.angle)))
        #if (abs(prop.angle)-abs(edge_ref.angle)) > 0.5/180*np.pi:
        #    continue
        x_offset_top = abs(b_box[0][0]+prop.x_at_height(0)-edge_ref.x_at_height(0))
        x_offset_bottom = abs(b_box[0][0]+prop.x_at_height(height)-edge_ref.x_at_height(height))
        x_offset_total = x_offset_top**2 + x_offset_bottom**2

        if x_offset_total < best_x_offset:
            best_edge = prop
            best_x_offset = x_offset_top
            print("\nUpdated best_x_offset")
        print("ref: {:4.0f}, {:4.0f} prop: {:4.0f}, {:4.0f} errors top: {:5.1f} bot: {:5.1f} total: {:5.1f}".format(edge_ref.x_at_height(0),
                                                                                                                    edge_ref.x_at_height(height),
                                                                                                                   b_box[0][0]+prop.x_at_height(0),
                                                                                                                   b_box[0][0]+prop.x_at_height(height),
                                                                                                                    x_offset_top,
                                                                                                                    x_offset_bottom,
                                                                                                                   x_offset_total))
        #print("dx: {:5.0f} dy: {:5.0f} angle: {:5.3f} x0: {:6.1f}".format(prop[0],prop[1],angle,x0))
        cv2.line(output, (l[0], l[1]), (l[2], l[3]), (0,0,255), 3, cv2.LINE_AA)
    cv2.line(output,(int(best_edge.x_at_height(0)),0),(int(best_edge.x_at_height(height)),height),(0,255,0),3,cv2.LINE_AA)

    plt.imshow(output)
    plt.show()

    img_mod = img.copy()
    pt1 = (b_box[0][0]+int(best_edge.x_at_height(0)),0)
    pt2 = (b_box[0][0]+int(best_edge.x_at_height(height)),height)
    cv2.line(img_mod,pt1,pt2,(0,255,0),3,cv2.LINE_AA)
    plt.imshow(img_mod)
    
plt.show()
```
